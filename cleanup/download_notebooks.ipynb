```python
import requests
import os

notebook_uris_file = "notebooks.txt"

with open(notebook_uris_file) as file:
    lines = [line.strip() for line in file.readlines()]

local_notebook_paths = []
local_folder = "official"
github_raw_prefix = "https://raw.githubusercontent.com/"
    
for url in lines:
    url = url.replace(
        "https://colab.research.google.com/github/",
        github_raw_prefix,
    )
    url = url.replace("blob/master", "master")
    request = requests.get(url, allow_redirects=True)

    if url.startswith(github_raw_prefix):
        file_name = url[len(github_raw_prefix):]
    else:
        file_name = url.split("/")[-1]
        
    file_name = f"{local_folder}/{file_name}"

    # Download
    print(f"Downloading {url} to {file_name}")
    
    # Make dirs
    if not os.path.exists(os.path.dirname(file_name)):
        try:
            os.makedirs(os.path.dirname(file_name))
        except OSError as exc: # Guard against race condition
            if exc.errno != errno.EEXIST:
                raise

    open(file_name, "wb").write(request.content)
    local_notebook_paths.append(f"cleanup/{file_name}")
```

```python
with open(f"cleanup_notebooks_local.txt", "w") as file:
    for element in local_notebook_paths:
        file.write(element + "\n")
```
